```python
dataset_name = "bigscience-data/roots_en_wikinews"
```

```python
ds = load_dataset(dataset_name, split="train")
```

```python
languages = []
for s in ds:
    match = re.search(r"'language': '([^']+)'", s['meta'])
    if match:
        language = match.group(1)
        languages.append(language)
print(len(languages), len(ds))
```

```python
set(languages)
```

```python
all_texts = []
for s in ds:
    all_texts.append(s["text"])
print(len(all_texts), len(ds))
```

```python
text = "\n\n".join(all_texts)
```

```python
len(text)
```

```python
chars = sorted(list(set(text)))
vocab_size = len(chars)
print(f"Vocab size: {vocab_size}")
```
